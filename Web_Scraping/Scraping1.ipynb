```python
#crear una varible raiz

root="https://www.football-data.co.uk/mmz4281/"
```

```python
#crear una lista de ligas

leagues=['E0','E1','E2','E3','EC']

frames=[]

#hacer loop en la lista, leer multiples csv y adjuntarlo 
for league in leagues:
    df=pd.read_csv(root+'2122'+'/'+league+'.csv',encoding='unicode_escape')
    frames.append(df)#con esto se va almacenar todos los dataframes, q estan en iteracion

    
```

```python
#df['season']=season esta es otra manera de inserta nueva columna, pero lo insertaria al final del dataset

# 3 ligas 
leagues=['E1','E2','E3']

frames=[]


#hacer looop a un rango de temporadas e insertar una nueva columna
for league in leagues:
    for season in range(15,21):#extraer la temporada 15 hasta la temporada 20

        df=pd.read_csv(root+str(season)+str(season+1)+'/'+league+'.csv')
        #crear una nueva columna que identifique a que temporada pertenece cada archivo .csv
        df.insert(1,'season',season)
        frames.append(df)

    
```

```python
#Crear un dicciionario con nombre original de liga como llave 

dic_countries={'Spanish La liga ':'SP1','Spanish Segunda Division':'SP2','German Bundesliga':'D1','English Premier League':'E0','English League 1':'E2','English League 2':'E3',}


```

```python
#OBTENER UN ELEMENTO DEL DICCIONARIO
dic_countries['Spanish La liga ']
```

```python
#MOSTRAR LA LLAVE
print('LLAVE ')
for league in dic_countries:
    print(league)
    
print("=====================================")
    
#MOSTRAR EL VALOR 
print('VALOR')
for league in dic_countries:
    print(dic_countries[league])
```

```python
#AQUI SE OBTENDRA UN DATAFRAME POR CADA LIGA Y ESE DATAFRAME VA A OBTENER LA DATA DE 5 O 6 TEMPORADAS

dict_historical_data={}


#hacer loop sobre llaves de diccionario
for league in dic_countries:
    frames=[]
    for season in range(15,21):#extraer la temporada 15 hasta la temporada 20

        df=pd.read_csv(root+str(season)+str(season+1)+'/'+dic_countries[league]+'.csv')
        #crear una nueva columna que identifique a que temporada pertenece cada archivo .csv
        df.insert(1,'season',season)
        frames.append(df)
    df_concat=pd.concat(frames)
    dict_historical_data[league]=df_concat

    

```
