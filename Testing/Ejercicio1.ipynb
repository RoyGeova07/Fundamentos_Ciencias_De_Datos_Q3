```python
import pandas as pd

datos={

    "Nombre": ["Roy", "Juan", "Pedro", "Ana", "Maria"],
    "Edad": [12, 15, 20, 18, 25],
    "Carrera": ["Ingenieria", "Medicina", "Arquitectura", "Sistemas", "Derecho"]

}
#sirve para crear un dataframe a partir de un diccionario
df=pd.DataFrame(datos)
df#sirve para mostrar el dataframe

```

```python
df["Nombre"]#sirve para mostrar una columna del dataframe
```

```python
df[["Nombre","Edad"]]#sirve para mostrar varias columnas del dataframe
```

```python
df["Edad"].mean()#sirve para calcular la media de una columna del dataframe
df["Edad"].max()#sirve para calcular el valor maximo de una columna del dataframe
df["Edad"].min()#sirve para calcular el valor minimo de una columna del dataframe
```

```python
df[df["Edad"]>18]#sirve para mostrar las filas que cumplen con una condicion
```

```python
df["Pais"]="Honduras"# sirve para agregar una nueva columna al dataframe
df# mostrar el dataframe con la nueva columna
```

```python
#cambiar el nombre de algunos paises de algunas personas variando entre otros paises
df.loc[df["Pais"]=="Honduras"]="Mexico"
df
```

```python
#cambiar el nombre de una columna
df.rename(columns={"Pais":"Pais de origen"}, inplace=True)#sirve para cambiar el nombre de una columna
df# mostrar el dataframe con el nuevo nombre de la columna
```

```python
#borrar columna del dataframe
df.drop(columns=["Pais de origen"],inplace=True)
df
```

```python
df["Edad x2"]=df["Edad"]*2#sirve para crear una nueva columna a partir de una columna existente
df
```

```python
df.sort_values(by="Edad")#ordenar el dataframe por una columna de manera ascendente por edad
df.sort_values(by="Edad",ascending=False)#ordenar el dataframe por una columna de manera descendente por edad
```

```python
df.describe()#sirve para mostrar un resumen estadistico del dataframe

#dara: 
#promedio
#máximo
#mínimo
#desviación estándar
#cuartiles
```

```python
import matplotlib.pyplot as plt

df["Edad"].plot(kind="bar")#sirve para graficar una columna del dataframe en un grafico de barras
plt.show()#sirve para mostrar el grafico
```

```python
df["Carrera"].value_counts().plot(kind="pie")#sirve para graficar una columna del dataframe en un grafico de pastel
plt.show()
```
